## Distribución Uniforme

Para generar números aleatorios uniformes entre dos valores dados usando scipy, podemos usar el siguiente código o el código que hemos
discutido en clase:

```python
from scipy.stats import uniform
import numpy as np

# Definir los límites inferior y superior de la distribución
lower_bound = 10  # límite inferior
upper_bound = 20  # límite superior

# Generar números aleatorios uniformemente distribuidos
# Utilizando un tamaño de muestra de 100 números
random_numbers = uniform.rvs(loc=lower_bound, scale=upper_bound-lower_bound, size=100)

# Imprimir los primeros 10 números generados
print("Ejemplo de números aleatorios generados:")
print(random_numbers[:10])

# También puedes generar un único número aleatorio
single_random_number = uniform.rvs(loc=lower_bound, scale=upper_bound-lower_bound)
print("\nUn solo número aleatorio generado:", single_random_number)
```

```python
import matplotlib.pyplot as plt
from scipy.stats import uniform

def generate_uniform_randoms(bottom, ceiling, count):
    random_numbers = uniform.rvs(loc=bottom, scale=ceiling-bottom, size=count)
    return random_numbers

```

```python
# Llamar a la función con los primeros parámetros: bottom=10, ceiling=15, count=100
random_numbers_1 = generate_uniform_randoms(10, 15, 100)
# Crear los histogramas
plt.figure(figsize=(12, 6))
# Histograma 1
plt.subplot(1, 2, 1)
plt.hist(random_numbers_1, bins=10, color='blue', edgecolor='black')
plt.title('Distribución Uniforme (bottom=10, ceiling=15)')
plt.xlabel('Valor')
plt.ylabel('Recuento')



```

```python
# Llamar a la función con los segundos parámetros: bottom=10, ceiling=60, count=1000
random_numbers_2 = generate_uniform_randoms(10, 60, 1000)

# Crear los histogramas
plt.figure(figsize=(12, 6))


# Histograma 2
plt.subplot(1, 2, 2)
plt.hist(random_numbers_2, bins=10, color='green', edgecolor='black')
plt.title('Distribución Uniforme (bottom=10, ceiling=60)')
plt.xlabel('Valor')
plt.ylabel('Recuento')

plt.tight_layout()
plt.show()
```

How are the two distributions different?

```python
# In summary, the main differences between the two distributions lie in the range of values generated and the number of samples taken. 
# Distribution 1 will have values tightly clustered within a smaller range, whereas Distribution 2 will have values spread out across a wider range. 
# The histograms visually represent these differences in distribution characteristics.
```

## Distribución Exponencial

1. Utilizando `numpy.random.exponential`, crea una función que devuelva una lista de números distribuidos exponencialmente con la media de 10. 

1. Utiliza la función para generar dos secuencias de números de tamaño 10 y 100.

1. 2. Traza las distribuciones como histogramas con el número de bins como 100.

El resultado debe ser como el siguiente:



```python
import numpy as np
import matplotlib.pyplot as plt

def generate_exponential_data(mean, size):
    return np.random.exponential(scale=mean, size=size)

# Generar datos exponenciales
data_size_10 = generate_exponential_data(mean=10, size=10)
data_size_100 = generate_exponential_data(mean=10, size=100)

# Crear histogramas
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(data_size_10, bins=100, color='blue', alpha=0.7)
plt.title('Distribución Exponencial (tamaño 10)')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

plt.subplot(1, 2, 2)
plt.hist(data_size_100, bins=100, color='green', alpha=0.7)
plt.title('Distribución Exponencial (tamaño 100)')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

¿En qué se diferencian ambas distribuciones?

La media cambia, por lo que la distribución también cambia. 

## Distribución Exponencial de datos reales

Supongamos que el tiempo que uno pasa en un banco se distribuye exponencialmente con media de 10 minutos (es decir, λ = 1/10). ¿Cuál es la probabilidad de que un cliente pase menos de quince minutos en el banco? 

Escribe un código en python para resolver este problema

```python
import numpy as np

# Parámetro lambda (tasa para la distribución exponencial)
lambd = 0.1  # Es equivalente a 1/10

# Tiempo límite en minutos
x = 15

# Calcular la probabilidad usando la CDF de la distribución exponencial
probability_less_than_15 = 1 - np.exp(-lambd * x)

print(f"La probabilidad de que un cliente pase menos de 15 minutos en el banco es: {probability_less_than_15:.4f}")
```
